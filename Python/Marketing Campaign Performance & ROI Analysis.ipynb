####   A/B Testing Problem Statement

## Business Problem

Companies invest in different marketing campaign types to acquire customers. However, it is unclear which campaign type generates better conversions.

## Objective

To determine whether Email marketing campaigns generate higher conversion rates than Social Media campaigns.

## A/B Test Design

- Group A (Control Group)
Email Campaigns

- Group B (Test Group)
Social Media Campaigns

Metric Used
Conversion Rate

## Hypothesis

- Null Hypothesis (H₀):
There is no significant difference in conversion rates between Email campaigns and Social Media campaigns.

- Alternative Hypothesis (H₁):
Email campaigns generate higher conversion rates than Social Media campaigns.

```python
ab_data = df[df['Campaign_Type'].isin(['Email','Social Media'])]
```

```python
email = ab_data[ab_data['Campaign_Type'] == 'Email']['Conversion_Rate']
social = ab_data[ab_data['Campaign_Type'] == 'Social Media']['Conversion_Rate']
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(email, social, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis: Email campaigns generate significantly higher conversion rates than Social Media campaigns.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference in conversion rates between Email and Social Media campaigns.")

####   A/B Test Result

The two-sample t-test results indicate that the p-value is greater than the significance level (α = 0.05). Therefore, we fail to reject the null hypothesis.

This suggests that there is no statistically significant difference in conversion rates between Email campaigns and Social Media campaigns.

 ## Business Interpretation

Although the average conversion rates of Email and Social Media campaigns may appear slightly different, the observed difference is not statistically significant and could be due to random variation in the data rather than a real effect.

##  Conclusion

- Based on the statistical test:

- Email campaigns do not significantly outperform Social Media campaigns.

- Social Media campaigns do not significantly outperform Email campaigns.

##  Business Recommendation

- Since both campaign types perform similarly in terms of conversion rate:

- Marketing budget can be allocated across both campaign types.

- Further analysis should focus on other factors, such as:

- marketing channel (Google Ads, Facebook, Instagram)

- target audience segments

- campaign duration

- These factors may provide more actionable insights for optimizing marketing performance.

##  A/B Testing Problem Statement

## Google Ads vs Facebook Campaigns


- Do Google Ads campaigns generate higher conversion rates than Facebook campaigns?

- Groups

- Group A: Google Ads campaigns
- Group B: Facebook campaigns

Metric

Conversion_Rate

##  Hypothesis

- Null Hypothesis (H₀):
There is no significant difference in conversion rates between Google Ads and Facebook campaigns.

- Alternative Hypothesis (H₁):
Google Ads campaigns generate higher conversion rates than Facebook campaigns.

```python
ab_data = df[df['Channel_Used'].isin(['Google Ads','Facebook'])]
```

```python
google_ads = ab_data[ab_data['Channel_Used'] == 'Google Ads']['Conversion_Rate']
facebook = ab_data[ab_data['Channel_Used'] == 'Facebook']['Conversion_Rate']
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(google_ads, facebook, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis: Google Ads campaigns generate significantly higher conversion rates than Facebook campaigns.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference in conversion rates between Google Ads and Facebook campaigns.")
```

####  A/B Test Result

The two-sample t-test produced a p-value greater than the significance level (α = 0.05). Therefore, we fail to reject the null hypothesis.

This indicates that there is no statistically significant difference in conversion rates between Google Ads and Facebook campaigns.

##  Business Interpretation

Although the average conversion rates for Google Ads and Facebook campaigns may differ slightly, the difference is not statistically significant and may be due to random variation in the data.

##  Conclusion

 Based on the statistical analysis:

- Google Ads campaigns do not significantly outperform Facebook campaigns.

- Facebook campaigns do not significantly outperform Google Ads campaigns.

##  Business Recommendation

Since both advertising platforms perform similarly in terms of conversion rate:

- Marketing budget can be allocated across both platforms.

- Further analysis should evaluate other factors such as:

-campaign type

- target audience

- customer segment

- campaign duration

These factors may provide more actionable insights for improving marketing performance.
